```python
from bs4 import BeautifulSoup
import os
import requests
```

```python
url= r"https://www.netflix.com/in/Login?nextpage=https%3A%2F%2Fwww.netflix.com%2Fbrowse"
```

```python
result= requests.get(url)
```

```python
result.status_code
```

```python
result.text
```

```python
soup = BeautifulSoup(result.text)
```

```python
img_tags = soup.find_all("img")
```

```python
img_tags
```

```python
for img_tag in img_tags:
    print(img_tag.get("src"))
```

```python
set_urls=set()
for img_tag in img_tags:
    set_urls.add(img_tag.get("src"))
```

```python
set_urls
```

```python
def save_img(url,dest):
    result= requests.get(url)
    if result.status_code==200:
            with open(dest,"wb") as photo:
                photo.write(result.content)
```

```python
if not os.path.exists("netflix_pics"):
    os.mkdir("netflix_pics")
for location in set_urls:
    total= location
    file_name=location.split("/")[-1]
save_img(total,"netflix_pics/"+file_name)    
```
